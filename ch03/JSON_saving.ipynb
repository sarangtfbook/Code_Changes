```
import tensorflow as tf
```

```
from tensorflow.keras.layers import Dense , Dropout , Conv2D , MaxPooling2D
```

```
# define model
model_1 = tf.keras.Sequential([
      Conv2D(32, (3, 3), activation = 'relu', padding = 'same', input_shape = (32, 32, 3)),
      Conv2D(32, (3, 3), activation = 'relu', padding = 'same'),
      MaxPooling2D((2, 2)),
      Dense(128, activation = 'relu'),
      Dense(10, activation = 'softmax')           
])
```

```
# model plot
tf.keras.utils.plot_model(model_1, 'multilayer_model.png', show_shapes=True)
```

```
# compile
model_1.compile(optimizer='adam',
              loss=tf.keras.losses.binary_crossentropy,
              metrics=['accuracy'])
```

```
# save model to JSON
json_model = model_1.to_json()
with open("model_1.json", "w") as json_file:
  json_file.write(json_model)
```

```
# reload model from JSON
from keras.models import model_from_json
file = open('model_1.json', 'r')
buffer = file.read()
file.close()
model =  tf.keras.models.model_from_json(buffer)
```

```
# plot loaded model for rechecking
tf.keras.utils.plot_model(model, 'multilayer_model.png', show_shapes=True)
```
